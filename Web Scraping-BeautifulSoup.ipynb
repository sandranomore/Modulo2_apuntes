```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
url = "https://www.decathlon.com/collections/deals?page=2"
res = requests.get(url)
print(res.status_code)
```

```python
# html-doc -> 
soup = BeautifulSoup(res.content, 'html.parser')
print(soup.prettify())
```

```python
lista_categorias = soup.find_all("span", {"class":"de-u-textMedium de-u-textShrink1 de-u-lg-textGrow1 de-u-textBold"})
lista_categorias
```

```python
lista_precios = soup.find_all("span", {"class":"js-de-ProductTile-currentPrice"})
lista_precios
```

```python
categorias = []

for i in lista_categorias:
    categorias.append(i.getText())

categorias
```

```python
# hecho como list comprehension
categorias = [i.getText() for i in lista_categorias]
categorias
```

```python
categorias = [i.replace('\n', "") for i in categorias]
```

```python
# quito los espacio
categorias = [i.strip() for i in categorias]
```

```python
precios = [i.getText() for i in lista_precios]
precios
```

```python
diccionario_decathlon = {}
diccionario_decathlon["categorias"] = categorias
diccionario_decathlon["precios"] = precios
diccionario_decathlon
```

```python
df = pd.DataFrame(diccionario_decathlon)
df.head(2)
```
